```python
from figrecipes.plotly.make_plots import PlotlyPlot
import pandas as pd
from IPython.display import Image
import numpy as np
```

*<b>In the below examples, we embedded the interactive plots (html files) in the notebook using "plot_mode='notebook'", however, Github is currently unable to render them (you can check the interactive plots when running this notebook on a Jupyter server). Thus, we have used "plot_mode='offline'", which opens and saves the interative plot as an html file, and embedded the static images here.*

## <b> Example of XY scatter plot </b>

```python
df = pd.read_csv('http://www.stat.ubc.ca/~jenny/notOcto/STAT545A/examples/gapminder/data/gapminderDataFiveYear.txt', 
                 sep='\t')
df2007 = df[df.year==2007]
print df2007.head(5)
```

```python
PlotlyPlot(plot_mode='offline', x_title='GDP per capita', y_title='Life expectancy', plot_title='Example XY plot', 
          filename='example_xy.html').\
xy_plot(x_col=df2007['gdpPercap'], y_col=df2007['lifeExp'], text=df2007['country'], size=df2007['gdpPercap'], 
        color=df2007['gdpPercap'])
```

```python
Image("example_xy.png")
```

## <b> Example #1 of heatmap plot </b>

```python
PlotlyPlot(plot_mode='offline', plot_title='Example heatmap 1', filename='example_heatmap1.html').\
    heatmap_plot(data=[[1, 20, 30, 50, 1], [20, 1, 60, 80, 30], [30, 60, 1, -10, 20]], 
    x_labels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'], y_labels=['Morning', 'Afternoon', 'Evening'])
```

```python
Image("example_heatmap1.png")
```

## <b> Example #1 of scatter matrix plot (using same colorscale for groups) </b>

```python
df_scatter = pd.DataFrame(np.random.randn(20, 4),
                         columns=['Column A', 'Column B', 'Column C', 'Column D'])
df_scatter['Fruit'] = pd.Series(['apple', 'apple', 'grape', 'apple', 'apple',
                                'grape', 'pear', 'pear', 'apple', 'pear',
                                'apple', 'apple', 'grape', 'apple', 'apple',
                                'grape', 'pear', 'pear', 'apple', 'pear'])
print df_scatter.head(5)
```

```python
PlotlyPlot(filename='example_scatterplot1.html').scatter_matrix(df_scatter, index_colname='Fruit')
```

```python
Image("example_scatterplot1.png")
```

```python
PlotlyPlot(filename='example_scatterplot2.html').scatter_matrix(df_scatter, index_colname='Fruit', diag_kind='box')
```

```python
Image("example_scatterplot2.png")
```
